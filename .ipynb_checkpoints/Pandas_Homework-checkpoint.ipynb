```python
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path
%matplotlib inline
```

# **Data Cleaning**

```python
# Reading whale returns
import csv
whale_returns_data= Path ("Resources/whale_returns.csv")
algo_returns_data= Path ("Resources/algo_returns.csv")
sp500_history_data= Path ("Resources/sp500_history.csv")


whale_returns_df= pd.read_csv (whale_returns_data, index_col= "Date", infer_datetime_format=True, parse_dates=True)
algo_returns_df= pd.read_csv (algo_returns_data, index_col= "Date", infer_datetime_format=True, parse_dates=True)
sp500_history_df= pd.read_csv (sp500_history_data, index_col= "Date", infer_datetime_format=True, parse_dates=True)

whale_returns_df.head()
```

```python
# Count nulls
whale_returns_df.isnull().sum()
```

```python
# Count nulls
whale_returns = whale_returns_df.dropna()
whale_returns.drop_duplicates()
whale_returns.head()
```

```python
whale_returns.isnull().sum()
```

```python
# Reading algorithmic returns
algo_returns_df.head()
```

```python
# Count nulls
algo_returns_df.isnull().sum()
```

```python
# Drop nulls
algo_returns = algo_returns_df.dropna()
algo_returns.drop_duplicates()
algo_returns.head()
```

```python
algo_returns.isnull().sum()
```

```python
# Reading S&P 500 Closing Prices
sp500_history_df.head()
```

```python
# Check Data Types
sp500_history_df.dtypes
```

```python
# Fix Data Types
sp500_history_df['Close'] = sp500_history_df['Close'].str.replace('$', '').astype("float")
sp500_history_df.dtypes
```

```python
# Calculate Daily Returns
sp500_daily_returns = sp500_history_df.pct_change()
sp500_daily_returns.sort_index()
sp500_daily_returns.head()
```

```python
# Drop nulls
sp500_daily_returns= sp500_daily_returns.dropna()
sp500_daily_returns.head()
```

```python
# Rename Column
sp500_daily_returns = sp500_daily_returns.rename(columns={
    "Close": "S&P 500",})
sp500_daily_returns.head()
```

```python
# Concatenate all DataFrames into a single DataFrame
combined_df_1 = pd.concat([whale_returns, algo_returns, sp500_daily_returns], axis="columns", join="inner")
combined_df_1.head()
```

# **Portfolio Analysis**

```python
# Plot daily returns
combined_df_1.plot()
```

```python
# Plot cumulative returns
cumulative_returns_df = (1 + combined_df_1).cumprod()
cumulative_returns_df.plot()
```

```python
# Box plot to visually show risk
cumulative_returns_df.plot(kind = "box", figsize = (20,5))
```

```python
# Daily Standard Deviations
# Calculate the standard deviation for each portfolio. 
# Which portfolios are riskier than the S&P 500?
std = cumulative_returns_df.std()
print(std)
```

```python
# Determine which portfolios are riskier than the S&P 500
risky_stocks = (std > std ['S&P 500'])
std ['Riskier than S&P 500'] = risky_stocks
std ['Riskier than S&P 500']
```

There are four stocks that are risker than the market.  The one's with higher risk would be Algo and Berkshire Hathaway Inc because they have the highest standard dev so there's a bit more volatility and uncertainty with these stocks. Then it would be Tiger Global Management LLC and Algo 2 since they are slightly above.

```python
# Calculate the annualized standard deviation (252 trading days)
annualized_variance = (combined_df_1.var()*252)
annualized_std = np.sqrt(annualized_variance)
annualized_std
```

# **Rolling Statistics**

```python
# Calculate and plot the rolling standard deviation for
# the S&P 500 and whale portfolios using a 21 trading day window
rolling_df= combined_df_1['BERKSHIRE HATHAWAY INC'].rolling(window=21).std().plot(figsize=(20,5))
```

```python
# Construct a correlation table
correlation = combined_df_1.corr()
correlation
```

```python
# Calculate Beta for a single portfolio compared to the total market (S&P 500)
# (Your graph may differ, dependent upon which portfolio you are comparing)
covariance = combined_df_1['BERKSHIRE HATHAWAY INC'].cov(combined_df_1['S&P 500'])
covariance
```

```python
variance = combined_df_1['BERKSHIRE HATHAWAY INC'].var()
variance
```

```python
berkshire_beta= covariance/variance
berkshire_beta
```

```python
rolling_covariance = combined_df_1['BERKSHIRE HATHAWAY INC'].rolling(window=60).cov(combined_df_1['S&P 500'])
rolling_variance = combined_df_1['S&P 500'].rolling(window=60).var()
```

```python
rolling_beta = rolling_covariance / rolling_variance
rolling_beta.plot(figsize=(20, 10), title='BERKSHIRE HATHAWAY INC Beta')
```

```python
# Calculate a rolling window using the exponentially weighted moving average. 
combined_df_1.ewm(halflife=21).std().plot(figsize=(20,10))
```

# **Sharpe Ratios**

```python
# Calculate annualized Sharpe Ratios
sharpe_ratios = (combined_df_1.mean() * 252) / (combined_df_1.std() * np.sqrt(252))
sharpe_ratios
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios.plot(kind="bar", title="Sharpe Ratios")
```

Algo 1 strategy seem to have outperform the market and the whales portfolio. Algo 2 strategy however didn't outperform the Berkshire Hathaway stock.

# **Portfolio Returns**

```python
# Read the first stock
goog_csv = Path("Resources/goog_historical.csv")
aapl_csv = Path("Resources/aapl_historical.csv")
cost_csv = Path("Resources/cost_historical.csv")

goog_df = pd.read_csv(goog_csv, index_col="Trade DATE", infer_datetime_format=True, parse_dates=True)
aapl_df = pd.read_csv(aapl_csv, index_col="Trade DATE", infer_datetime_format=True, parse_dates=True)
cost_df = pd.read_csv(cost_csv, index_col="Trade DATE", infer_datetime_format=True, parse_dates=True)

goog_df.head()
```

```python
# Read the second stock
aapl_df.head()
```

```python
# Read the third stock
cost_df.head()
```

```python
aapl_df.rename( columns={"Symbol": "AAPL1","NOCP":"AAPL"},inplace = True )
aapl_df.drop(columns=['AAPL1'], axis=1, inplace= True)
```

```python
cost_df.rename( columns={"Symbol": "COST1","NOCP":"COST"},inplace = True )
cost_df.drop(columns=['COST1'], axis=1,inplace = True)
```

```python
goog_df.rename( columns={"Symbol": "GOOG1","NOCP":"GOOG"},inplace = True )
goog_df.drop(columns=['GOOG1'], axis=1, inplace = True)
```

```python
# Concatenate all stocks into a single DataFrame
combined_df = pd.concat([aapl_df,cost_df,goog_df], axis='columns', join='inner')
combined_df.head()
```

```python
combined_df.sort_index(inplace=True)
combined_df.head()
```

```python
# Drop Nulls
combined_df.dropna()
combined_df.head()
```

```python
daily_returns = combined_df.pct_change().dropna()
daily_returns.head()
```

## **Calculate the weighted returns**

```python
# Calculate weighted portfolio returns
weights = [1/3, 1/3, 1/3]
portfolio_returns = daily_returns.dot(weights)

custom_portfolio = pd.DataFrame(portfolio_returns)
custom_portfolio.head()
```

## **Join the portfolio returns**

```python
# Add your "Custom" portfolio to the larger dataframe of fund returns
combined_df_new = pd.concat([portfolio_returns, combined_df_1], axis='columns', join='inner')
combined_df_new.rename(columns= {0: "Custom"},inplace = True )
combined_df_new.dropna(inplace=True)
combined_df_new.sort_index()
combined_df_new.head()
```

# **Risk Analysis**

```python
# Risk
combined_df_new.std()
annualized_variance_new = (combined_df_new.var()*252)
annualized_std_new = np.sqrt(annualized_variance_new)
annualized_std_new
```

```python
combined_new.rolling(window=21).std().plot(figsize=(15,8))
```

```python
# Annualized Sharpe Ratios
new_sharpe_ratios = (combined_df_new.mean() * 252) / ((combined_df_new).std() * np.sqrt(252))
new_sharpe_ratios
```

```python
# Visualize the sharpe ratios as a bar plot
new_sharpe_ratios.plot(kind="bar", title="Sharpe Ratios")
```

```python
# Create a correlation analysis
correlation = combined_df_new.corr()
correlation
```

```python
# Beta
rolling_variance = combined_df_new['Custom'].rolling(window=21).var()
rolling_covariance = combined_df_new['Custom'].rolling(window=21).cov(combined_df_new['S&P 500'])
rolling_beta = rolling_covariance / rolling_variance
rolling_beta.plot(figsize=(20, 10), title='Custom Portfolio Beta')
```
